```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
df_2011 = pd.read_csv('/content/drive/MyDrive/DATA6300/articles2011-oct2021.csv')
```

```python
from tqdm.notebook import tqdm
```

```python
response = requests.get('https://recalls-rappels.canada.ca/en/alert-recall/undeclared-allergens-2-greek-gals-products')
soup = BeautifulSoup(response.text, 'html.parser')

information_grid = soup.find_all('div','recall-alert-body')

for tag in information_grid:
    tag_table = tag.find_all('dt',class_='paddingNone')
    print(tag_table)

con_list = []
for con in tag_table:
    con_list.append(con.text)
print(con_list)
```

```python
starting_date2 = []
communication2 = []
alert2 = []
sub_cat2 = []
hazard_class2 = []
source2 = []
firm2 = []
distribution2 = []
extent2 = []
ref2 = []
for report_url in tqdm(df_2011['Report URL']):
    response = requests.get(report_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    information_grid = soup.find_all('div','recall-alert-body')

            #print(information_grid)

    for info in information_grid:
        table = info.find_all('dd',class_='paddingNone')

    for tag in information_grid:
        tag_table = tag.find_all('dt',class_='paddingNone')

    missing_list = []
    tag_con_list = []
    tag_tx_list = []


    for con,tx in zip(tag_table,table):
        tag_con_list.append(con.text)
        tag_tx_list.append(tx.text)
    print(tag_con_list)
    print(tag_tx_list)

    for i in con_list:
        if i in tag_con_list:
            missing_list.append(i)
        else:
            missing_list.append('None')
    print(missing_list)


    info_list = []
    for i in range(len(missing_list)):
        if missing_list[i] == 'None':
            tag_tx_list.insert(i,'None')
    print(tag_tx_list)
    starting_date2.append(tag_tx_list[0])
    communication2.append(tag_tx_list[1])
    alert2.append(tag_tx_list[2])
    sub_cat2.append(tag_tx_list[3])
    hazard_class2.append(tag_tx_list[4])
    source2.append(tag_tx_list[5])
    firm2.append(tag_tx_list[6])
    distribution2.append(tag_tx_list[7])
    extent2.append(tag_tx_list[8])
    ref2.append(tag_tx_list[9])
```

```python
report_df = pd.DataFrame({'Starting Date':starting_date2,'Type of Communication':communication2,'Alert':alert2,'Sub-Category':sub_cat2,'hazard Classification':hazard_class2,'source':source2,'firm':firm2,'distribution':distribution2,'extent':extent2,'ref. no':ref2})
```

```python
report_df.to_csv('2011-2021_allergen_new_report_final.csv', index = False, encoding='utf-8')
```
