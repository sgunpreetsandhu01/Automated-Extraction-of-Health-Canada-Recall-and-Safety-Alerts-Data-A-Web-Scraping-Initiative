```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://recalls-rappels.canada.ca/en/alert-recall/food-recall-warning-allergen-cache-cuisine-brand-pro-licious-protein-pancake-waffle'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup)
```

```python
content = soup.find_all('div','ar-additional-info ar-section')
```

```python
print(content)
```

```python
details = []
for i in content:
    identity=[]
    date = i.find('time').text
    alert = i.find(class_='field field--name-field-recall-type field--type-entity-reference field--label-hidden field--item').text.strip()
    category = i.find(class_='field field--name-field-category field--type-entity-reference field--label-hidden field--items').text.strip()
    companies = i.find(class_='field field--name-field-companies field--type-text-long field--label-hidden field--item').text.strip()
    publisher = i.find(class_='field field--name-field-organization field--type-entity-reference field--label-hidden field--item').text.strip()
    audience = i.find(class_='field field--name-field-who-this-is-for field--type-entity-reference field--label-hidden field--items').text.strip()
    audience = i.find(class_='field field--name-field-who-this-is-for field--type-entity-reference field--label-hidden field--items').text.strip()
    recall = i.find(class_='field field--name-field-hazard-type field--type-entity-reference field--label-hidden field--items').text.strip()
    for j in i.find_all(class_='field field--name-node-id field--label-inline'):
        identity.append(j.text.strip())
    id_no = [k.split('\n') for k in identity]
#     identity = i.find(class_='field field--name-node-id field--label-inline').text.strip()
    print(date)
    print(alert)
    print(category)
    print(companies)
    print(publisher)
    print(audience)
    print(recall)
    print(id_no)
    temp2=i.find('div','field--item')
    if temp2==None:
        continue
    print(temp2)
print(details)
```
