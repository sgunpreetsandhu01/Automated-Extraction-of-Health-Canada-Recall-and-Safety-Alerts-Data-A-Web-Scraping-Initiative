```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('allergen_new_report_final_2011.csv')
```

```python
df.head()
```

```python
df['Sub-Category'].loc[8].split(',')
```

```python
subs = []
for i in df['Sub-Category']:
  subs.append(i.split(','))
```

```python
new_sub = set()
for sub in subs:
  for i in sub:
    new_sub.add(i)
```

```python
new_sub = list(new_sub)
```

```python
new_sub
```

```python
df.head()
```

```python
# Returns the list that contains info about the particular sub-category, if present '1' is appended, else '0' is appended into the list.
def cat_encode(sub):
  l = []
  for i in range(len(df['Sub-Category'])):
    if sub in df['Sub-Category'].iloc[i]:
      l.append(1)
    else:
      l.append(0)
  return l
```

```python
# Looping through the unique list of sub-categories and encoding it using the function defined earlier
for sub in new_sub:
  df[sub] = cat_encode(sub)
```

```python
df.head()
```

```python
distribution = []
for dis in df['distribution']:
  distribution.append(dis.split(','))
```

```python
new_dis = set()
for dis in distribution:
  for i in dis:
    new_dis.add(i)
```

```python
new_dis = list(new_dis)
```

```python
new_dis
```

```python
df['distribution'].value_counts()
```

```python
# Returns the list that contains info about the particular distribution, if present '1' is appended, else '0' is appended into the list.
def distribution_encode(dis):
  l = []
  for i in range(len(df['distribution'])):
    if dis in df['distribution'].iloc[i]:
      l.append(1)
    else:
      l.append(0)
  return l
```

```python
# Looping through the unique list of sub-categories and encoding it using the function defined earlier
for dis in new_dis:
  df[dis] = distribution_encode(dis)
```

```python
df.head()
```

```python
df.to_csv('report_encoded.csv',index=False,encoding='utf-8')
```

```python
def count_distribution(dis):
  count = 0
  for i in range(len(df['distribution'])):
    if dis in df['distribution'].iloc[i]:
      count+=1
  return dis,count
```

```python
#intializing `distribution_count` dataframe
distribution_count = pd.DataFrame(columns=['distribution','count'])
for i in range(len(new_dis)):
  distribution_count.loc[i] = count_distribution(new_dis[i])
```

```python
distribution_count
```

```python
#Sorting `distribution_count` in descending order
distribution_count.sort_values(by=['count'],ascending=False,inplace=True)
```

```python
plt.figure(figsize=(40,15))
plt.title('Distribution')
plt.xlabel('Provinces')
plt.ylabel('Count')
plt.bar(distribution_count['distribution'],distribution_count['count'])
plt.show()
```
