```python
import requests
from bs4 import BeautifulSoup

def extract_info(url):
    while url:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        # Find the 'next' button link to go to the next page
        next_button = soup.find('a', {'title': 'Go to next page'})
        base_url = 'https://recalls-rappels.canada.ca/en/search/site'
        return (url := base_url + next_button['href'] if next_button else None)

# Starting URL
start_url = 'https://recalls-rappels.canada.ca/en/search/site?search_api_fulltext=allergen&archived=1&f%5B0%5D=recall_type%3A4%27&page=%2C0%2C0'
extract_info(start_url)
```

```python
link = []
link.append(start_url)
url = extract_info(start_url)
link.append(url)
while url:
    url = extract_info(link[-1])
    print(url)
    link.append(url)
    
```

```python
import pandas as pd
links = pd.DataFrame(link,columns=['links'])
```

```python
links.to_csv('links.csv', index = False, encoding='utf-8')
```

```python
report_urls = []
dates = []
for l in link: 
    response = requests.get(l)
    soup = BeautifulSoup(response.text, 'html.parser')

    recall_base = 'https://recalls-rappels.canada.ca'
    article_grid = soup.find_all('div','search-result views-row')

    for article in article_grid:
        report_url = recall_base + article.find('a')['href']
        date = article.find(class_='ar-type').text.strip().split('|')[1]
        print(report_url)
        print(date)
        report_urls.append(report_url)
        dates.append(date)
        
```

```python
article_df = pd.DataFrame({'Report URL':report_urls,'Dates':dates})
```
