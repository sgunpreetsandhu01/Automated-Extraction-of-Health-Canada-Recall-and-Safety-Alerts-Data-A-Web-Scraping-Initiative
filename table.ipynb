```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
df_2011 = pd.read_csv('/content/drive/MyDrive/DATA6300/articles2011-oct2021.csv')
df_2011.head()
```

```python
from tqdm.notebook import tqdm
brand_name = []
common_name = []
result_common_name = []
st_date = []
ref_url = []
d=0
for report_url in tqdm(df_2011['Report URL']):
    response = requests.get(report_url)
    print(report_url)
    st_date.append(df_2011['Dates'].loc[d])
    d+=1
    ref_url.append(report_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    table_grid = soup.find_all('table','mrgn-tp-sm')

    if len(table_grid) > 0:
      td_tag = []
      for i in table_grid:
          td_tag = i.find_all('td')

      td_text = []

      for td in td_tag:
          td_text.append(td.text.strip())

      items = []
      for it in range(1,len(td_text),5):
          items.append(td_text[it])

      items_list = "-".join(items)
      common_name.append(items_list)

      pro_items = []
      for tt in range(0,len(td_text),5):
          pro_items.append(td_text[tt])

      pro_items_list = "-".join(pro_items)
      brand_name.append(pro_items_list)

    else:
      common_name.append('None')
      brand_name.append('None')



print(common_name)
print(brand_name)
```

```python
table_df = pd.DataFrame({'Starting Date':st_date,'Ref URL':ref_url,'Brand Names':brand_name,'Common Names':common_name})
```

```python
table_df.to_csv('new_table.csv',index=False,encoding='utf-8')
```
